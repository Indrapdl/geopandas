```python
# Importing libraries
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# importing and reading population data 
population = pd.read_csv('population.csv')
population.head()
```

```python
# Filter population data of syangja district only
population_syangja = population[population['District']=='Syangja']
```

```python
population_syangja
```

```python
 # importing and reading nepal shapefile 
shapefile = gpd.read_file('D:\\RMU Project\\Local Unit\\local_unit.shp')
shapefile
```

```python
# Filter for the local level within 'Syangja' district
district = 'SYANGJA'  # Make sure the case matches your data
syangja = shapefile[shapefile['DISTRICT'].str.upper() == district]
```

```python
syangja
```

```python
# plotting shapefile geometry 
syangja.plot(cmap = 'jet',column = 'GaPa_NaPa')
plt.title('Syangja District-Local Levels')
plt.show()
```

```python
# identifying coordinate system of shapefile
syangja.crs
```

```python
# projecting to required coordinate system(utm 44 in this case)
syangja_proj = syangja.to_crs(32644)
syangja_proj.crs
```

```python
# Calculating area of each local units
syangja_proj['local_unit_Area(km^2)'] = syangja_proj.area/1000000
syangja_proj
```

```python
population_syangja.shape
```

```python
syangja_proj.shape
```

```python
# Add a new field common field to both dataframe and geodataframe to faciliate attribute join
population_syangja['join_index'] = range(1, len(population_syangja) + 1)

syangja_proj['join_index'] = range(1, len(syangja_proj) + 1)
```

```python
syangja_proj
```

```python
population_syangja
```

```python
# Joining population data(pandas dataframe) with shapefile geometry data(geopandas geodataframe) 
syangja_joined = syangja_proj.merge(population_syangja,on = 'join_index')
syangja_joined
```

```python
# Dropping duplicate and not useful columns
syangja_joined.drop(['GaPa_NaPa', 'Type_GN','DISTRICT','join_index'], axis=1,inplace = True)
```

```python
syangja_joined
```

```python
# Calculating population density
syangja_joined['popn_density(per_sq_km)']=syangja_joined['Total population']/syangja_joined['local_unit_Area(km^2)']
syangja_joined
```

```python
# Plotting the population density map with  legend
fig, ax = plt.subplots(1, 1, figsize=(12, 12))
syangja_joined.plot(
    column='popn_density(per_sq_km)', 
    ax=ax, 
    legend=True,
    legend_kwds={
        'label': "Population Density (people per sq. km)",
        'orientation': "horizontal",
        'shrink': 0.6  # Shrinks the size of the legend if needed
    },
    cmap='viridis'  
)

# Adding labels of local level names  using apply and ax.annotate()
syangja_joined.apply(lambda x: ax.annotate(text=x['Local Level Name'], xy=x['geometry'].centroid.coords[0],ha='center', fontsize=7, color='black'), axis=1)
                                      
# Adding title and axis labels
ax.set_title('Population Density Map of Syangja', fontdict={'fontsize': 20, 'fontname': 'Times New Roman'})
ax.set_xlabel('Easting (m)')
ax.set_ylabel('Northing (m)') 


# Display the plot
plt.show()
```

```python
#Saving map to file
output_path= 'D:\\RMU project\\Syangja_population_density.png'
fig.savefig(output_path, dpi=250) 
plt.close()
```
